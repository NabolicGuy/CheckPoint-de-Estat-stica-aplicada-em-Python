1)Luke Skywalker solicitou ao seu time de cientistas de dados para que analisassem os dados financeiros da Resistência (aliança contra o Império) para saber se terão recursos suficientes para combater seus inimigos. A Resistência, cansada de vender cursos de “como dominar seu sabre de luz” na internet, lançou uma marca de camisetas chamada Jedi que bombou em alguns anos, mas também gerou muitas despesas devido aos fretes intergaláticos. Após analisarem os dados de faturamento e despesas da Resistência ao longo dos anos, vocês descobriram as seguintes funções: ● Faturamento em função do tempo: F(t) = 2t³ - 5t² + 4t - 1 ● Despesas em função do tempo: D(t) = t³ - 3t² +2t +1


```python
#a) L(t) = F(t)-D(t)
import numpy as np
from scipy.integrate import quad

def faturamento(t):
    return 2*t**3 - 5*t**2 + 4*t - 1

def despesas(t):
    return t**3 - 3*t**2 + 2*t + 1

def lucro(t):
    return faturamento(t) - despesas(t)

#b)
a = 3
b = 6

area, _ = quad(lucro, a, b)

print(f"A função do lucro é: L(t) = t³ - 2t² + 2t - 2")
print(f"A área sob a curva do lucro entre {a} e {b} anos é: {area}")



#C)
import matplotlib.pyplot as plt

def lucro(t):
    return t**3 - 2*t**2 + 2*t - 2

t = np.linspace(3, 6, 600)

lucros = lucro(t)

plt.figure(figsize=(8, 6))
plt.plot(t, lucros, label='Lucro(t)')

plt.title('Gráfico do Lucro da Resistência')
plt.xlabel('Anos')
plt.ylabel('Lucro')
plt.grid(True)
plt.legend()
plt.fill_between(t, lucros, color='skyblue', alpha=0.4)

plt.show()
```

d) Essa área representa o total acumulado de lucros ou prejuízos ao longo do período. Ao calcularmos essa área, encontramos um valor aproximado de 198.75 unidades de área. Como essa área é positiva, isso significa que, ao longo dos anos, a Resistência teve mais momentos de lucro do que de prejuízo, ou que os lucros foram maiores do que os prejuízos.
Portanto, podemos concluir que a Resistência teve um lucro líquido entre 3 e 6 anos. Isso ocorre porque o faturamento superou as despesas ao longo do período, resultando em um saldo positivo.

2) Darth Vader, após ter sido ressuscitado por uma máquina de
última geração da Resistência e ter feito um transplante capilar na Turquia
graças ao cupom de desconto de um blogueiro espacial, decidiu colocar sua

quantidade de lixo espacial jogado na galáxia pelas naves do Império, ele
decidiu montar uma estação de coleta e reciclagem de materiais de dois tipos
de lixo, o lixo orgânico (O) e o lixo reciclável (R), durante uma semana.
Com isso, poderia prover mais recursos para que a Resistência possa operar
de forma sustentável e ao mesmo tempo ajudar a galáxia. Com a ajuda do
seu grupo de cientistas de dados foi possível descobrir que a quantidade de
cada tipo de lixo coletado ao longo do tempo pode ser representada pelas
funções O(t) e R(t), respectivamente, onde t é o tempo em dias, e a
quantidade de lixo é medida em toneladas.
A função que descreve a quantidade de lixo orgânico O(t) coletado por dia
pela estação da Resistência é dada por:
O(t) = 2t² - 5t + 3
A função que descreve a quantidade de lixo reciclável R(t) coletado por dia é
dada por:
R(t) = 2t³ - 4t² + 6
Darth Vader quer calcular a área entre os gráficos de coleta do lixo
orgânico e do lixo reciclável durante os dias t=2 e t=6, para entender a
diferença total acumulada de material coletado entre os dois tipos de lixo
nesse período. Para isso, ele solicitou que vocês

```python
#a)
from scipy.integrate import quad

def O(t):
    return 2*t**2 - 5*t + 3

def R(t):
    return 2*t**3 - 4*t**2 + 6

def area_diferenca(t):
    return abs(O(t) - R(t))

t_min, t_max = 2, 6
area_total, _ = quad(area_diferenca, t_min, t_max)
print(f"A área entre os gráficos de coleta de lixo orgânico e lixo reciclável são de: {area_total:.2f} toneladas")
```

```python
#B)
t = np.linspace(2, 6, 100)
O = 2 * t**2 - 5 * t + 3
R = 2 * t**3 - 4 * t**2 + 6


plt.plot(t, O, label="Lixo Orgânico O(t)", color="purple")
plt.plot(t, R, label="Lixo Reciclável R(t)", color="green")

plt.xlabel("Tempo (dias)")
plt.ylabel("Quantidade de lixo (toneladas)")
plt.legend()
plt.title("Coleta de Lixo Orgânico e Reciclável")
plt.grid()
plt.show()
```

3)Buscando ajudar os planetas menos favorecidos pelo Império, a
Resistência Jedi está analisando o índice de Gini de alguns locais da galáxia.
A princípio, coletaram os dados do planeta Lothal com a ajuda de sua equipe
de Cientistas de Dados e, através de uma regressão polinomial, vocês
obtiveram a função que descreve a relação entre distribuição de renda em
percentual e a quantidade de habitantes alienígenas em percentual.
O modelo descoberto foi g(x) = 0.22x² + 0.06x + 0.001, onde x é a quantidade
percentual da população. Com isso, a Resistência pediu para que vocês:



```python

def g(x):
    return 0.22 * x**2 + 0.06 * x + 0.001


x = np.linspace(0, 1, 1000)

y = g(x)


y = y / y[-1]

area_lorenz = np.trapz(y, x)


gini = 1 - 2 * area_lorenz


print(f"O índice de Gini é: {gini:.4f}")
```
